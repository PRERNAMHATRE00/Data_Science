```python
import matplotlib.pyplot as plt
import numpy as np

values = np.array([5, 6, 3, 7, 2])
names = np.array(["A", "B", "C", "D", "E"])
colors = np.array(["Green", "Blue", "Red", "Yellow", "Orange"])

plt.barh(names, values, color=colors,height = 0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
ages=[2,5,8,18,20,26,45,66,80,90,75]

range=[0,100]
bins=10

plt.hist(ages,bins,range,
         color='purple',histtype='bar',edgecolor='black')

plt.xlabel('age')
plt.ylabel("no.of people")
plt.title=("Histogram")
```

```python
import matplotlib.pyplot as plt
ages=[2,5,8,18,20,26,45,66,80,90,75]

range=[0,100]
bins=10

plt.hist(ages,bins,range,
         color='purple',histtype='bar',edgecolor='black',alpha=0.500)

plt.xlabel('age')
plt.ylabel("no.of people")
plt.title=("Histogram")

```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6, 9]  # Removed the last element from x
y = [0, 1, 4, 9, 11, 12, 16, 25]
plt.scatter(x, y, label="stars", s=100, color="blue", marker="*")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Scatter plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
activities=["eat",'sleep','work','play']
slices=[3,7,8,6]
colours=["r",'y','b','g']
x = [0, 1, 2, 3, 4, 5, 6, 9]  # Removed the last element from x
y = [0, 1, 4, 9, 11, 12, 16, 25]
plt.pie(slices,labels=activities,colors=colours,
        startangle=90,shadow=True, explode=(0,0,0.5,0),radius=1.2, autopct='01.1f88')
plt.legend()

plt.show()
```
